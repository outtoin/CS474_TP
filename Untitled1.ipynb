```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
with open ('../results/results0.4.txt', 'r') as f:
    result = f.readlines()
```

```python
result = [line.split('\n')[0] for line in result]
results = [line.split(':') for line in result if line.split(':')[1] == ' Accuracy']
```

```python
result[0].split(':')[1]
```

```python
results[0][0].split('/')[1] 
```

```python
taskA = np.array([line for line in results if 'taskA' in line[0].split('/')[1]])
taskB = np.array([line for line in results if 'taskB' in line[0].split('/')[1]])
taskC = np.array([line for line in results if 'taskC' in line[0].split('/')[1]])
```

```python
taskA[:, 0] = list(map(lambda s: s[15:-6], taskA[:, 0]))
taskB[:, 0] = list(map(lambda s: s[15:-6], taskB[:, 0]))
taskC[:, 0] = list(map(lambda s: s[15:-6], taskC[:, 0]))
```

```python
dfA = pd.DataFrame(taskA)
dfA = dfA.drop(1, axis=1)
dfA.columns = ['a', 'Accuracy']
df1 = pd.DataFrame(dfA.a.str.split(',').tolist(), columns=['num', 'k', 'roughness'])
dfA = pd.concat([df1, dfA.Accuracy], axis=1)
dfA = dfA.infer_objects()
dfA = dfA.apply(pd.to_numeric, errors='ignore')
```

```python
dfB = pd.DataFrame(taskB)
dfB = dfB.drop(1, axis=1)
dfB.columns = ['a', 'Accuracy']
df1 = pd.DataFrame(dfB.a.str.split(',').tolist(), columns=['num', 'k', 'roughness'])
dfB = pd.concat([df1, dfB.Accuracy], axis=1)
dfB = dfB.apply(pd.to_numeric, errors='ignore')
```

```python
dfA
```

```python
dfC = pd.DataFrame(taskC)
dfC = dfC.drop(1, axis=1)
dfC.columns = ['a', 'Accuracy']
df1 = pd.DataFrame(dfC.a.str.split(',').tolist(), columns=['num', 'k', 'roughness'])
dfC = pd.concat([df1, dfC.Accuracy], axis=1)
dfC = dfC.apply(pd.to_numeric, errors='ignore')
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.set_ylim([0.2, 0.55])
plt.title('Accuracy at TaskA', fontsize=20)
plt.xlabel('SVQ size', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.xticks(rotation=90)
sns.pointplot(taskA[:, 0], taskA[:, 2].astype(np.float), linestyles=["-"])
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.set_ylim([0.35, 0.45])
plt.title('Accuracy at TaskA', fontsize=20)
plt.xlabel('SVQ size', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
sns.pointplot(x='num', y='Accuracy', hue='k', data=dfA, palette='hls')
```
